```python
'''Tugas kelompok Farischa Makay & Akhner Frischa : Mencari nilai terkecil dari
3 input user''' 

def switch():
 
    option = int(input("Please choose number (1 or 2). 1 for Faricha coding and 2 for Akhner coding: "))
 
 
    if option == 1:
        value1 = input(("Please enter number 1 :\n"))
        value2 = input(("Please enter number 2 :\n"))
        value3 = input(("Please enter number 3 :\n"))
        
        
        if value1 < value2 and value1 < value3:
          print(f'{value1} is the lowest number')
        elif value2 < value1 and value2 < value3 :
          print(f'{value2} is the lowest number')
        elif value3 < value2 and value3 < value1 :
          print(f'{value3} is the lowest number')
        elif value1 == value2 and value1 != value3 :
          print(f'{value1} is the lowest number')
        elif value1 == value3 and value1 != value2 :
          print(f'{value1} is the lowest number')
        elif value2 == value3 and value2 != value1 :
          print(f'{value2} is the lowest number')
        else:
          print("Your numbers have same result.  Please change your number.")

 
    elif option == 2:
        print("Program Menentukan Nilai Terkecil Dari 3 Angka")
        data = []
        jum = 0

        for i in range(3):
          temp = int(input("Masukkan Angka ke-%d: " % (i+1)))
          data.append(temp)
          jum += data[i]
    
        print('Nilai Terkecil Sekarang Adalah :', min(data))
 
    else:
        print("Incorrect option")
 
 
 
switch()
 
```
